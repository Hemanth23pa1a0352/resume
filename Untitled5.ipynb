```python
import numpy as np
import google.generativeai as client

client.configure(api_key='AIzaSyCuOd4qMQnDV7cofZ-WhYXJkDJaxNB14K4')

def cosine_similarity(vec1, vec2):
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

sentences = [
    "I love machine learning",
    "The weather is very hot today",
    "AI models use neural networks",
    "I enjoy studying artificial intelligence"
]

query = "learning AI"

# Embed query
query_response = client.embed_content(
    model="text-embedding-004",
    content=query
)
query_emb = np.array(query_response['embedding'])

scores = []

for sent in sentences:
    sent_response = client.embed_content(
        model="text-embedding-004",
        content=sent
    )
    sent_emb = np.array(sent_response['embedding'])

    score = cosine_similarity(query_emb, sent_emb)
    scores.append((sent, score))

# Sort by similarity
scores.sort(key=lambda x: x[1], reverse=True)

for s, sc in scores:
    print(s, "->", sc)
```
